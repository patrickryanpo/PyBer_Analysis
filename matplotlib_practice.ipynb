```python
# Object oriented approach
fig, ax = plt.subplots()
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*10)
ax.scatter(y_axis, x_axis, c="skyblue",s=y_axis, alpha=.8, edgecolors="k", linewidth = 2, label='Boston')
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0,50)
ax.invert_yaxis()
```

```python
# Create pie chart
plt.pie(y_axis, labels=x_axis)
```

```python
# Increase size of Pie Chart
plt.subplots(figsize=(8,8))

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange","gold", "indianred", "tomato","mistyrose"]
# Explode values 
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, colors=colors, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```
